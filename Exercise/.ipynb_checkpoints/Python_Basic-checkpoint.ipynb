```python
# 중첩 루프 사용하기 - 별 찍기
for i in range(5): # 세로 처리 코드
    for j in range(5): # 가로 처리 코드
        if i >= j:
            print('*', end='')
        if j == 4:
            print()
            
print()

for i in range(5):
    for j in range(5):
        if i == j:
            print('*', end = "")
        else:
            print(' ', end = "")
        if j == 4:
            print()

print()

for i in range(5):
    for j in range(5):
        # i == 0, j == 4 / i == 1, j == 3, 4 / i == 2, j == 2, 3
        if i + j < 4:
            print(' ', end = "")
        else:
            print('*', end = "")
        if j == 4:
            print()

print()

for i in range(5):
    for j in reversed(range(5)):
        if j > i:
            print(' ', end = '')
        else:
            print('*', end = '')
    for j in range(5):
        if j < i:
            print('*', end = '')
    print()
# for i in range(5):
#     for j in range(5):
#         print('j: ', j)
#     for j in reversed(range(5)):
#         print('j2:', j)
```

```python
# 재귀호출 (재귀함수)
def hello():
    print('Hello World!')
    hello()
    
def hello(count):
    if count == 0:
        return
    print('Hello World!', count)
    count -= 1
    hello(count)

hello(5)
```

```python
# Factorial
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n-1)
```

```python
# Max, Min, Sum 함수 만들기
a = [38, 21, 53, 62, 19]
def maximum(list):
    largest = list[0]
    for i in list:
        if largest < i:
            largest = i
    return largest

print(maximum(a))

def sum_(list):
    sum__ = 0
    for i in list:
        sum__ += i
    return sum__

print(sum_(a))
```

```python
# Class Method
# classmethod, staticmethod
class Person:
    count = 0 # class attribute
    def __init__(self, name):
        self._name = name
        Person.count += 1
    
    @classmethod
    def count_person(cls):
        print('{0} 명 생성되었습니다.'.format(cls.count))
        
    @classmethod
    def create_person(cls, name):
        person = cls(name)
        return person
    
    @staticmethod
    def add(a, b):
        print(a + b)

James = Person(name = 'James')
Maria = Person(name = 'Maria')
Person.create_person('Eungi')
Person.count_person()
Person.add(1, 2)

# abstract method - abstract class cannot make instance
from abc import *
class StudentBase(metaclass = ABCMeta):
    @abstractmethod
    def study(self):
        pass
    
    @abstractmethod
    def go_to_school(self):
        pass

class Student(StudentBase):
    def study(self):
        print('study')
    def go_to_school(self):
        print('go_to_school') 
        # If no: Can't instantiate abstract class Student with abstract method go_to_school
        
# class inheritance
class Universe(object):
    def __init__(self):
        print('Universe __init__')
        self.owner = 'god'
    def greeting(self):
        print('Hello')

class Earth(Universe):
    def __init__(self):
        super().__init__()
        print('Earth __init__')
        self.people = 'people'
    def greeting(self):
        super().greeting()
        print('Hallo')

        
earth = Earth()
print(earth.people)
print(earth.owner) # 만약 super() 없었다면 오류
earth.greeting()
```

```python
# 예외 만들기
class NotThreeMultipleError(Exception):
    def __init__(self):
        super().__init__('3의 배수가 아닙니다. 잘못 입력했습니다.')

# 예외 처리
try:
    mul_3 = int(input('3의 배수를 입력하세요: '))
    if mul_3 % 3 != 0:
        raise NotThreeMultipleError()
    print(mul_3)
except Exception as e:
    print('예외가 발생했습니다.', e)
else:
    print('3의 배수가 정확히 맞습니다.') # 예외가 발생하지 않을 경우
finally:
    print('종료합니다.') # 항상 출력
```

```python
# 클로저 및 데코레이터
# https://engineer-mole.tistory.com/181
def foo(arg):
    x = 10
    print(locals())
foo(20)

def outer(x):
    def inner():
        print(x)
    return inner
foo1 = outer(11) # x = 11 고정인수를 얻음
foo2 = outer(12) # x = 12 고정인수를 얻음


def outer(some_func):
    def inner():
        print('before some_func')
        ret = some_func()
        return ret + 1 # inner는 some_func을 실행한 결과에 1을 더한 값을 리턴하는 함수
    return inner

def foo():
    return 1
foo = outer(foo)
foo() # decorator 는 foo + 어떠한 처리라고 이해할 수 있음.

# 어떠한 좌표의 객체를 유지하는 라이브러리를 사용하고자한다. 
# 이 객체는 x와 y의 좌표 쌍을 가지고 있지만, 아쉽게도 덧셈이나 나눗셈 등 수학 처리 기능은 없다. 
# 그러나 우리가 이 라이브러리를 이용해서 대량의 계산 처리를 할 필요가 있어 라이브러리의 소스를 개편해야하는
# 반갑지 않은 상황이다. 

class coordinate(object):
    def __init__(self, x, y):
        self.x = x
        self.y = y
def add(a, b):
    # a, b is instance
    return coordinate(a.x + b.x, a.y + b.y)

one = coordinate(100, 200)
two = coordinate(200, 300)
three = add(one, two)
print(three.x, three.y)

# 좌표계는 반드시 (0, 0) 이상이어야 할 경우, 
# 좌표를 체크 해주는 decorator를 작성하여 add, sub를 decorate 해주자
def wrapper(func):
    def check(a, b):
        print('Execute Check')
        if a.x < 0 or a.y < 0:
            a = coordinate(a.x if a.x >= 0 else 0, a.y if a.y >= 0 else 0)
        if b.x < 0 or b.y < 0:
            b = coordinate(b.x if b.x >= 0 else 0, b.y if b.y >= 0 else 0)
        ret = func(a, b)
        if ret.x < 0 or ret.y < 0:
            ret = coordinate(ret.x if ret.x >= 0 else 0, ret.y if ret.y >= 0 else 0)
        return ret
    return check

add = wrapper(add)
added_instance = add(one, two)
print(added_instance.x, added_instance.y)

@wrapper
def sub(a, b):
    return coordinate(a.x - b.x, a.y - b.y)

subtracted_instance = sub(one, two)
print(subtracted_instance.x, subtracted_instance.y)


# 함수의 인수를 출력해주는 데코레이터
def print_parameter(func):
    def inner(*arg, **kwargs):
        print(arg, kwargs)
        ret = func(*arg, **kwargs)
        return ret
    return inner

@print_parameter
def foo_(*arg, **kwargs):
    print(arg[0] * kwargs['x'])

foo_(1,2,3,4,6, x = 2, y = 5, z = 9)
```
