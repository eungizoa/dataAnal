The cut method for pandas data splits the dataset into bins. There are a number of arguments for the method. The following code creates equal sized bins. 
The method value_counts returns a frequency table.

```python
# Important - histogram 선행작업 - cut method
binnedPopulation = pd.cut(state['Population'], 10)
print(binnedPopulation)
# value_counts() method를 통해 count출력
print(binnedPopulation.value_counts()) 
```

```python
# Table 1.5
binnedPopulation.name = 'binnedPopulation'
print(type(binnedPopulation), type(state))
df = pd.concat([state, binnedPopulation], axis = 1)
df = df.sort_values(by = 'Population')
# df = df.reindex(range(len(df)))
df
```

```python
df.groupby(['binnedPopulation']).count()
```

```python
# Important
# histogram으로 나눠진 각 부분을 dataframe으로 자세하게 나타내기
groups = []
for group, subset in df.groupby(by='binnedPopulation'):
    groups.append({
    'BinRange' : group,
    'Count' : len(subset),
    'States' : ', '.join(subset['Abbreviation']) # join함수
    })
print(groups)
# Important
# 판다스 선언이 리스트 안에 여러 딕셔너리가 담겨있는 방식으로도 될 수 있다.
pd.DataFrame(groups)
```

```python
from matplotlib.collections import EllipseCollection
from matplotlib.colors import Normalize

def plot_corr_ellipses(data, figsize=None, **kwargs):
    ''' https://stackoverflow.com/a/34558488 '''
    # DataFrame을 2차원 행렬로 dump하기
    M = np.array(data)
    print(M.shape[1])
    if not M.ndim == 2:
        raise ValueError('data must be a 2D array')
    # fig, ax 정리: https://hwi-doc.tistory.com/entry/matplotlib-%EC%99%84%EB%B2%BD-%EC%A0%95%EB%A6%AC
    fig, ax = plt.subplots(1, 1, figsize=figsize, subplot_kw={'aspect':'equal'})
    # x축, y축 lim 설정
    ax.set_xlim(-0.5, M.shape[1] - 0.5)
    ax.set_ylim(-0.5, M.shape[0] - 0.5)
    # y축 반전
    ax.invert_yaxis()

    # xy locations of each ellipse center
    # reshape: https://rfriend.tistory.com/345
    xy = np.indices(M.shape)[::-1].reshape(2, -1).T

    # set the relative sizes of the major/minor axes according to the strength of
    # the positive/negative correlation
    w = np.ones_like(M).ravel() + 0.01
    # raval() - 1차원으로 전환
    h = 1 - np.abs(M).ravel() - 0.01
    # ??
    a = 45 * np.sign(M).ravel()
    # sign 함수 리턴 -1 if x < 0, 0 if x==0, 1 if x > 0 . nan은 nan 입력에 대해 반환됩니다.
    
    ec = EllipseCollection(widths=w, heights=h, angles=a, units='x', offsets=xy,
                           norm=Normalize(vmin=-1, vmax=1),
                           transOffset=ax.transData, array=M.ravel(), **kwargs)
    ax.add_collection(ec)

    # if data is a DataFrame, use the row/column names as tick labels
    if isinstance(data, pd.DataFrame):
        ax.set_xticks(np.arange(M.shape[1]))
        ax.set_xticklabels(data.columns, rotation=90)
        ax.set_yticks(np.arange(M.shape[0]))
        ax.set_yticklabels(data.index)
        # set_xticks, set_xticklabels vs set_xlim
        # https://m.blog.naver.com/PostView.naver?isHttpsRedirect=true&blogId=jung2381187&logNo=220408589520
    return ec, ax

m, ax = plot_corr_ellipses(etfs.corr(), figsize=(5, 4), cmap='bwr_r')
cb = fig.colorbar(m, ax=ax)
cb.set_label('Correlation coefficient')

plt.tight_layout()
plt.show()
```

```python
kc_tax = pd.read_csv(KC_TAX_CSV)
kc_tax
```

```python
# 수치형 변수 대 수치형 변수 시각화 (육각형 구간과 등고선)
kc_tax0 = kc_tax.loc[(kc_tax['TaxAssessedValue'] < 750_000) &
          (kc_tax['SqFtTotLiving'] > 100) &
          (kc_tax['SqFtTotLiving'] < 3_500), :]
kc_tax0.shape
```

```python
# 육각형 구간: hexagonal binning
# scatter plot처럼 점으로 표시하는 대신, 기록값을 육각형 모양의 구간들로 나누고 각 구간에 포함된 기록값의 개수에 따라
# 색깔을 표시

# sharex 에 관하여: https://wikidocs.net/159927
ax = kc_tax0.plot.hexbin(x = 'SqFtTotLiving', y = 'TaxAssessedValue', gridsize = 30,
                        sharex = False, figsize = (5,4))
ax.set_xlabel('Finished Square Feet')
ax.set_ylabel('Tax-Assessed Value')

plt.tight_layout()
plt.show()
```

```python
# 등고선: contour plot
# 등고선은 두 변수로 이루어진 지형에서의 등고선을 의미. 꼭대기로 갈수록 밀도는 높아진다.
fig, ax = plt.subplots(figsize = (4,4))
ax = sns.kdeplot(data = kc_tax0, x = 'SqFtTotLiving', y = 'TaxAssessedValue', ax=ax)
# 해당 결과는 several minutes이 소요되므로, sample을 통해 나타내도 전체적인 shape는 유지될 것.
# Important. 아래와 같은 방식으로 sample을 추출할 수 있음.
# ax = sns.kdeplot(data=kc_tax0.sample(10000), x='SqFtTotLiving', y='TaxAssessedValue', ax=ax)
ax.set_xlabel('Finished Square Feet')
ax.set_ylabel('Tax-Assessed Value')
plt.show()
```

```python
# 범주형 변수 대 범주형 변수
lc_loans = pd.read_csv(LC_LOANS_CSV)
print(lc_loans.size)
lc_loans.head(10)
```

```python
crosstab = lc_loans.pivot_table(index = 'grade', columns = 'status',
                                aggfunc = lambda x: len(x), margins = True)
crosstab
```

```python
df = crosstab.loc['A':'G', :].copy()
df.loc[:, 'Charged Off':'Late'] = df.loc[:, 'Charged Off':'Late'].div(df['All'], axis = 0)
df['All'] = df['All'] / sum(df['All'])
perc_crosstab = df
perc_crosstab
```

```python
# 다변수 시각화하기: conditioning (조건화)
# 다수의 그래프 그리기: https://mindscale.kr/course/python-visualization-basic/grid/
zip_codes = [98188, 98105, 98108, 98126]
# Important
# isin method
kc_tax_zip = kc_tax0.loc[kc_tax0['ZipCode'].isin(zip_codes)]
kc_tax_zip
```

```python
def hexbin(x, y, color, **kwargs):
    cmap = sns.light_palette(color, as_cmap = True)
    plt.hexbin(x, y, gridsize = 25, cmap = cmap, **kwargs)

g = sns.FacetGrid(kc_tax_zip, col = 'ZipCode', col_wrap = 2)
g.map(hexbin, 'SqFtTotLiving', 'TaxAssessedValue',
     extent = [0, 3500, 0, 700000])
g.set_axis_labels('Finished Squrare Feet', 'Tax-Assessed Value')
g.set_titles('Zip code {col_name: .0f}')

plt.tight_layout()
plt.show()
```
