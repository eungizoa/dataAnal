```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

from tqdm import tqdm
tqdm.pandas()

import warnings
warnings.filterwarnings('ignore')
```

```python
df_attr = pd.read_csv('/Users/cho-eungi/Desktop/market_entry_attributes_cnt.csv')
```

```python
def softmax(Series):
    scaled_Series = (Series - Series.min()) / (Series.max() - Series.min())
    if not scaled_Series.any():
        scaled_Series.replace(np.nan, 1 / len(scaled_Series), inplace = True)
    
    exp_Series = np.exp(scaled_Series)
    sum_exp_Series = np.sum(exp_Series)
    result = exp_Series / sum_exp_Series
    return result

df_attr['soft_max'] = df_attr.groupby(['market_id']).progress_apply(
                                                                    lambda df: softmax(df['attribute_sum'])
                                                     ).reset_index().set_index('level_1').drop('market_id', axis=1)

sum_test = df_attr.groupby('market_id')['soft_max'].aggregate('sum')
print('all of the market has sum of soft_max == 1?', sum_test.sum() == len(sum_test))
```

## EDA

```python
'''
1) number of entries by market
'''

entry_cnt = df_attr.groupby('market_id')['entry_id'].aggregate(len)

ax = entry_cnt.plot(
    figsize = (30, 5), title = 'Number of Entries by Market',
    xlabel = 'Market ID', ylabel = 'Number of entries',
    color = '#00bcd4', rot = 90
)
```

```python
'''
2) Number of unique values in column 'attribute_sum' by market
'''
unique_cnt = df_attr.groupby('market_id').aggregate({'attribute_sum' : 'nunique'})
ax = unique_cnt.plot(
    figsize = (30, 5), title = 'Number of unique values in column "attribute_sum" by Market',
    xlabel = 'Market ID', ylabel = 'Number of unique values',
    color = '#00bcd4'
)
```

## Method

```python
market_ids = unique_cnt.loc[unique_cnt['attribute_sum'] == 10].index.to_list()
sample_df = df_attr[
                df_attr['market_id'].isin(np.random.choice(market_ids, 1))
            ].sort_values(by = 'soft_max')

sample_df
```

```python
count_freq = pd.DataFrame(
                [*sample_df.groupby('soft_max').apply(
                                        lambda series : {
                                            'soft_max' : series.iloc[0, 3],
                                            'count' : len(series)}
                )]
             )
count_freq['cum_count'] = count_freq['count'].cumsum()
count_freq['weight_group'] = pd.qcut(count_freq['cum_count'], 
                                     10, 
                                     labels = [i/10 for i in range(1, 11)], retbins = False)
soft_max_grouped = count_freq[['soft_max', 'weight_group']]
soft_max_grouped = pd.merge(sample_df, soft_max_grouped, on = 'soft_max')
```

```python
soft_max_grouped['weight_group'] = soft_max_grouped['weight_group'].astype('float64')
soft_max_grouped['weight_scaled'] = soft_max_grouped['weight_group'] * (1 / soft_max_grouped['weight_group'].sum())
```
